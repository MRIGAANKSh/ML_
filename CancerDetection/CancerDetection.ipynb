```python
# prompt: open data.csv using df.head

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Breast Cancer Predication System
df=pd.read_csv('/content/data.csv')

df.head()
```

# Spliting the Data in x and y.

```python
X=df.drop(['id','diagnosis'],axis=1)
```

```python
X
```

```python
y=df.diagnosis.values
```

# Training and Testing Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

# FEature Scailing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

```

```python
X_train
```

Model building

```python
from sklearn.neighbors import KNeighborsClassifier
#initalizethe class first
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)
```

```python
ypredknn=classifier.predict(X_test)
```

```python
ypredknn
```

```python
y_test
```

```python
#Classification Matrice to check how the model behave

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,ypredknn)*100,2)
```
